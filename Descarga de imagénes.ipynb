```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.1f}'.format
```

```python
df = pd.read_csv('../Img_genoma/Copia de obra.csv', skiprows=1)
```

```python
df_url = df[['TOKEN', 'URL_IMG_1', 'URL_IMG_2', 'URL_IMG_3', 'URL_IMG_4', 'URL_IMG_5']]
```

```python
name = []
for i in range(len(df_url)):
    temp1 = str(i) + '-genoma-' + (df.TOKEN.iloc[i])
    name.append(temp1)
```

```python
df_url['name'] = name
```

```python
df_url = df_url.fillna(0)
```

```python
df_url.head()
```

```python
columns = ['URL_IMG_1', 'URL_IMG_2','URL_IMG_3', 'URL_IMG_4', 'URL_IMG_5']
for col in columns:    
    for url in range(len(df_url)):
        if not df_url.iloc[url][col] == 0:
            try:
                url_imagen = df_url.iloc[url][col] # El link de la imagen
                nombre_local_imagen = str(df_url.name.iloc[url] + '-' + str(col) + '.jpg')
                print(nombre_local_imagen)
                folder = 'Img/'
                imagen = requests.get(url_imagen).content
                with open(folder + nombre_local_imagen, 'wb') as handler:
                    handler.write(imagen)
            except:
                print('ERROR' + str(nombre_local_imagen))
        else:
            pass
```
